```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bhp.csv")
df
```

```python
df.head()
```

```python
df.price_per_sqft.describe()
```

```python
df.price_per_sqft.quantile(0.25)
```

```python
percentile_001, percentile_999 = df.price_per_sqft.quantile([0.001,0.999])
percentile_001, percentile_999
```

```python
df2 = df[(df.price_per_sqft >= percentile_001) & (df.price_per_sqft <= percentile_999)]
df2
```

```python
df2.describe()
```

```python
std = df2.price_per_sqft.std()
std
```

```python
mean = df2.price_per_sqft.mean()
mean
```

```python
df3 = df2[(df.price_per_sqft >= mean - 4 * std) & (df.price_per_sqft <= mean + 4 * std)]
df3
```

```python
plt.hist(df3.price_per_sqft, rwidth = .95)
plt.title("Property Prices in Bangalore")
plt.xlabel("Price per Square Feet")
plt.ylabel("No. of properties")
```

```python
df2["zscore"] = (df2.price_per_sqft - df2.price_per_sqft.mean())/df2.price_per_sqft.std()
df2
```

```python
df4 = df2[(df2.zscore >= -4) | (df2.zscore <= 4)]
df4
```
