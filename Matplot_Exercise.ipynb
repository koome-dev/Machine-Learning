```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
values = np.arange(len(company))
values
```

```python
plt.xticks(values,company)
plt.bar(values, revenue)
plt.title("US Stocks")
plt.ylabel("Revenue in US Billions")
plt.savefig("ustocks.pdf", bbox_inches="tight",pad_inches=0.5)
```
