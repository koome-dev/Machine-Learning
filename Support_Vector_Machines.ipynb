```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df["target"] = digits.target
df.head()
```

```python
digits.target_names
```

```python
df["number name"] = df.target.apply(lambda x: digits.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(["target", "number name"], axis=1)
```

```python
y = df.target
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=20)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```
