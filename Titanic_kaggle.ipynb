```python
new_df = df.fillna(value={"Age":28})
new_df
```

```python
titanic_df = new_df[["Survived", "Pclass", "Sex", "Age", "Fare"]]
titanic_df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
titanic_df["le_Sex"] = le_sex.fit_transform(titanic_df["Sex"])
titanic_df.head()
```

```python
inputs = titanic_df.drop(["Survived", "Sex"], axis="columns")
inputs
```

```python
target = titanic_df[["Survived"]]
target
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```

```python
from sklearn.svm import SVC
model_svc = SVC()
```

```python
model_svc.fit(inputs,target)
```
